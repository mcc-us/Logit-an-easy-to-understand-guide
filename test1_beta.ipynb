```python
import matplotlib.pyplot as plt
import numpy as np
import math
# %matplotlib notebook
%matplotlib inline

x = np.linspace(-6, 6, 21)
y = 1/(1+np.exp(-1*x))

plt.figure(figsize=(10, 7))  # 10 is width, 7 is height
plt.plot(x, y, 'g', marker='D')  # go=green dots
plt.title('A Simple Logistic Curve', fontsize=24)
plt.xlabel('X', fontsize=24)
plt.ylabel('Probability', fontsize=24)
plt.xlim(-6, 6)
plt.ylim(-0.1, 1.1)

# plt.plot(x, y=1, color='go')  # go=green dots

plt.axhline(y=0.5, color='r', linestyle='-.')
plt.axhline(y=0, color='b', linestyle='dotted')
plt.axhline(y=1, color='b', linestyle='dotted')
plt.axvline(x=0, color='g')
plt.text(-5, 0.52, 'Y=0 If Pr(x) < 0.5', fontsize=18)
plt.text(1.5, 0.52, 'Y=1 If Pr(x) >= 0.5', fontsize=18)
plt.text(-1.9, 0.92, 'Inflection Pt, x=0', fontsize=18)
plt.yticks(np.arange(0, 1.3, 0.5), fontsize=14)
plt.xticks(np.arange(-6, 6, 2), fontsize=14)
plt.show()
```
